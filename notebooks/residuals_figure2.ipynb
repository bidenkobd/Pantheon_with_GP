# Define functions for distance calculations

```python
#for individual chain pionts

def h_z_arr(z,*args):
    H0,om,pwr = args
    h = ( H0 * ( om * (1+z)**3 + (1-om) ) ** 0.5 ) ** pwr
    return h


def da_arr(arr ,pwr = -1):
    z, H0,om = arr
    if z==0: d=0
    else: d = intgr.quad(h_z_arr,0.,z,args=(H0,om,pwr))[0]*(astropy.constants.c.value/1000.)/(1+z)

    return d

#from mean values

def h_z(z,*args):
    H0,om,pwr = args
    ol  = 1 - om
    h = ( H0 * ( om * (1+z)**3 + ol ) ** 0.5 ) ** pwr
    return h
def da(z, H0,om,pwr = -1):
    d = intgr.quad(h_z,0.,z,args=(H0,om,pwr))[0]*(astropy.constants.c.value/1000.)/(1+z)
    return d
```

# Calculate redshift binned data for Pantheon+

```python
xxx = dict()
xxx['mean'] = []
xxx['std'] = []
xxx['z'] = []
    
mb = []
for i in data['zHD']: mb.append(5*np.log10(da(i,panp_nogp.getMeans()[0],panp_nogp.getMeans()[1])*(1+i)**2) + 25 + panp_nogp.getMeans()[2])
    
zmin = 0.01    

for i in range(len(lc_parameters_alt['zcmb'])):
    if i == 39: 
        zmax = 3.
    else: 
        zmax = (lc_parameters_alt['zcmb'][i]+lc_parameters_alt['zcmb'][i+1])/2.
    zsel = (data['zHD'].values>zmin)&(data['zHD'].values<zmax)
    sttt = DescrStatsW(data['m_b_corr'].values[zsel]-np.array(mb)[zsel],data['m_b_corr_err_DIAG'].values[zsel]**-2)
    xxx['mean'].append(sttt.mean)
    xxx['std'].append(np.power(np.sum(np.power(data['m_b_corr_err_DIAG'].values[zsel],-2)),-1./2.))
    xxx['z'].append(np.mean(data['zHD'].values[zsel]))
    zmin = zmax
zz_long_p = np.append(np.append(min(data['zCMB'].values),xxx['z']),max(data['zCMB'].values))
```

# Create arrays with redshifts and mcmc paramer points

Pantheon

```python
zz = lc_parameters_alt['zcmb'].values
```

```python
zz_long = np.append(np.append(min(lc_parameters['zcmb'].values),zz),max(lc_parameters['zcmb'].values))
```

```python
arg_arr=[]
for i in zz_long:
    arg_arr.append(np.array([i*np.ones(len(pan_nogp.getParams().H0_gp)),pan_nogp.getParams().H0_gp,pan_nogp.getParams().omegam_gp]).T)
```

Pantheon+

```python
arg_arr_p=[]
for i in zz_long_p:
    arg_arr_p.append(np.array([i*np.ones(len(panp_nogp.getParams().H0_gp)),panp_nogp.getParams().H0_gp,panp_nogp.getParams().omegam_gp]).T)
np.array(arg_arr_p);
```

# calculate distances 

Pantheon

```python
da_shoes_lcdm = []
for i in range(zz_long.size):
    da_shoes_lcdm.append(list(map(da_arr,arg_arr[i])))
da_shoes_lcdm = np.array(da_shoes_lcdm)
```

Pantheon+

```python
da_shoes_lcdm_p = []
for i in range(len(arg_arr_p)):
    da_shoes_lcdm_p.append(list(map(da_arr,arg_arr_p[i])))
da_shoes_lcdm_p = np.array(da_shoes_lcdm_p)
```

# get confidence limits

Pantheon

```python
norm=[]
for i in range(len(da_shoes_lcdm[0])):
    norm.append(pan_nogp.get_norm(where=i))
mean=[]
u1=[]
u2=[]
d1=[]
d2=[]
err1=0.16
err2=0.025
for i in range(0,zz_long.size):
    stt = DescrStatsW(5*np.log10(da_shoes_lcdm[i]*(1+zz_long[i])**2)+25+pan_nogp.getParams().M_gp, weights=norm, ddof=0)
    mean.append(stt.mean)
    u1.append(float(stt.quantile(err1)))
    u2.append(float(stt.quantile(err2)))
    d1.append(float(stt.quantile(1-err1)))
    d2.append(float(stt.quantile(1-err2)))
mean=np.array(mean)
u1=np.array(u1)
u2=np.array(u2)
d1=np.array(d1)
d2=np.array(d2)
```

Pantheon+

```python
norm_p=[]
for i in range(len(da_shoes_lcdm_p[0])):
    norm_p.append(panp_nogp.get_norm(where=i))
mean_p=[]
u1_p=[]
u2_p=[]
d1_p=[]
d2_p=[]
err1=0.16
err2=0.025
for i in range(len(zz_long_p)):
    if np.isnan(zz_long_p[i]): print('oh no')
    else:    
        stt = DescrStatsW(5*np.log10(da_shoes_lcdm_p[i]*(1+zz_long_p[i])**2)+25+panp_nogp.getParams().M_gp, weights=norm_p, ddof=0)
        mean_p.append(stt.mean)
        u1_p.append(float(stt.quantile(err1)))
        u2_p.append(float(stt.quantile(err2)))
        d1_p.append(float(stt.quantile(1-err1)))
        d2_p.append(float(stt.quantile(1-err2)))
mean_p=np.array(mean_p)
u1_p=np.array(u1_p)
u2_p=np.array(u2_p)
d1_p=np.array(d1_p)
d2_p=np.array(d2_p)
```

# Comparing Pantheon and Pantheon+ data to infered Pantheon+ cosmology

```python
f,ax = plt.subplots(1,2,dpi=400,sharey=True,gridspec_kw={'wspace':0.1});
f.set_figwidth(12)
f.set_figheight(5)

c1 = '#3820FF'
c2 = '#FF2020'
zlg10 = np.log10(zz_long)
zlg10p = np.log10(zz_long_p)
## confidence limits from infered parameters

#Pantheon
ax[0].fill_between(zlg10,u1-mean,d1-mean,alpha=0.1,lw=0,color=c1,label=None)
ax[0].fill_between(zlg10,u2-mean,d2-mean,alpha=0.1,lw=0,color=c1,label=None)

#Pantheon+
ax[1].fill_between(zlg10p,u1_p-mean_p,d1_p-mean_p,alpha=0.1,lw=0,color=c2,label=None)
ax[1].fill_between(zlg10p,u2_p-mean_p,d2_p-mean_p,alpha=0.1,lw=0,color=c2,label=None)

#zero residual line
ax[0].hlines(0,min(zlg10p),max(zlg10p),lw=1,color=c1,label = '$\Lambda$CDM reconstruction')
ax[1].hlines(0,min(zlg10p),max(zlg10p),lw=1,color=c2,label = '$\Lambda$CDM reconstruction')


#binned residual data with statistical errors 
ax[1].errorbar(np.log10(xxx['z']),xxx['mean'], yerr=xxx['std'],
               fmt='.',ms=4,elinewidth=0.5, lw=0.,label = 'Pantheon+ binned data',c=c2)
ax[0].errorbar(np.log10(zz),lc_parameters_alt['mb']-mean[1:-1], 
               yerr=lc_parameters_alt['dmb'],  fmt='.',ms=4,elinewidth=0.5, lw=0.,label = 'Pantheon binned data',c=c1)

# plot limits
ax[0].set_xlim(min(zlg10),max(zlg10))
ax[1].set_xlim(min(zlg10),max(zlg10))

# vertical dashed lines at z=0.08 and z=0.15
ylim = ax[0].get_ylim()
ax[0].vlines(np.log10(0.08),ylim[0],ylim[1],linestyles = '--', color='k', alpha = 0.5,lw=1.)
ax[0].vlines(np.log10(0.15),ylim[0],ylim[1],linestyles = '--', color='k', alpha = 0.5,lw=1.)
ax[0].set_ylim(ylim)

ylim = ax[1].get_ylim()
ax[1].vlines(np.log10(0.08),ylim[0],ylim[1],linestyles = '--', color='k', alpha = 0.5,lw=1.)
ax[1].vlines(np.log10(0.15),ylim[0],ylim[1],linestyles = '--', color='k', alpha = 0.5,lw=1.)
ax[1].set_ylim(ylim)

#legends
ax[0].legend()
ax[1].legend()

#labels
ax[0].set_ylabel('$m_b-{m_b}_{\Lambda CDM}$')
ax[0].set_xlabel('log $z$')
ax[1].set_xlabel('log $z$')

f.savefig('pan_res_vs_panp_res.pdf',bbox_inches='tight')
```

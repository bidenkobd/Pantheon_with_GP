# Loading everything


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pn
%matplotlib inline
```

```python
font = {'size'   : 16, 'family':'DejaVu Sans'}
matplotlib.rc('font', **font)
plt.rcParams["figure.figsize"] = [4.0,4.0]
plt.rcParams["figure.dpi"] = 150
cmap = plt.cm.get_cmap('hot')
```

# Load Pantheon data

```python
a=np.loadtxt('../data/sys_full_long.txt')[1:]
b=pn.read_csv('../data/lcparam_def.txt',delimiter = ' ')
a= a.reshape((1048,1048))
cov = a+np.diag(b['dmb'].values**2)
```

# Load Pantheon+ data

```python
data = pn.read_table(r'../data/Pantheon+SH0ES.dat',sep=' ')
covp = pn.read_table(r'../data/Pantheon+SH0ES_STAT+SYS.cov',sep=' ')
covp = np.reshape(covp.values,(1701,1701))
```

# Compuiting separation between Pantheon sample SNe

Pantheon

```python
dist = np.zeros((len(b['zcmb'].values),len(b['zcmb'].values)))
for j in range(len(b['zcmb'].values)):
    for i in range(len(b['zcmb'].values)):
        dist[i,j] = np.abs(b['zcmb'].values[j]-b['zcmb'].values[i])
```

Pantheon+

```python
distp = np.zeros((len(data['zCMB'].values),len(data['zCMB'].values)))
for j in range(len(data['zCMB'].values)):
    for i in range(len(data['zCMB'].values)):
        distp[i,j] = np.abs(data['zCMB'].values[j]-data['zCMB'].values[i])
```

# Plotting redshift-sorted covariance matrix

```python
# Pantheon

# getting sorting arguments 
srtarr = np.argsort(b['zcmb'])

f,ax = plt.subplots(1,1)
f.set_dpi(400)

# plotting full covariance matrix
im=ax.imshow((((cov)[srtarr]).T[srtarr])[::-1],cmap='Greys',
          vmin=-0.001,
          vmax=0.001)
# plotting shaded region for \delta z < 0.18
ax.imshow((((dist<0.18)[srtarr]).T[srtarr])[::-1]-0.3,cmap='hot',alpha=((((dist<0.18)*0.1)[srtarr]).T[srtarr])[::-1],vmin=0.,vmax=1.),#)#,

cb = f.colorbar(im,fraction =.1,shrink=0.87,format= '%.4f')

cb.ax.tick_params(labelsize=6) 
ax.axis('off')
f.savefig('cov_pan.pdf',bbox_inches = 'tight')
```

```python
# Pantheon+

f,ax = plt.subplots(1,1)
f.set_dpi(400)

# plotting full covariance matrix
im=ax.imshow((covp).T[::-1],cmap='Greys',
          vmin=-0.001,
          vmax=0.001)
# plotting shaded region for \delta z < 0.18
ax.imshow(((distp<0.18).T)[::-1]-0.3,cmap='hot',alpha=(((distp<0.18)*0.1).T)[::-1],vmin=0.,vmax=1.),#)#,

cb = f.colorbar(im,fraction =.1,shrink=0.87,format= '%.4f')

cb.ax.tick_params(labelsize=6) 
ax.axis('off')
f.savefig('cov_panp.pdf',bbox_inches = 'tight')
```

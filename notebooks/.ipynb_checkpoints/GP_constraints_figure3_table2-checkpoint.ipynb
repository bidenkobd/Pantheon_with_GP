```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from statsmodels.stats.weightstats import DescrStatsW
import os
import pandas as pn
import scipy
from getdist.mcsamples import loadMCSamples
import astropy.constants
import scipy.integrate as intgr
import getdist.plots as gdplt
%matplotlib inline
```

```python
c1 = '#3820FF'
c2 = '#FF2020'
```

```python
font = {'size'   : 16, 'family':'DejaVu Sans'}
matplotlib.rc('font', **font)
plt.rcParams["figure.figsize"] = [4.0,4.0]
plt.rcParams["figure.dpi"] = 150
cmap = plt.cm.get_cmap('hot')
```

# Load mcmc chains with infered cosmological posteriors

Pantheon

```python
pan_nogp =loadMCSamples(os.path.abspath('../chains/pan_nogp/pan_nogp_023_mcmc_comas_2nditer_v_final'),settings={'ignore_rows':0.7})
pan_nogp.addDerived(pan_nogp.getParams().H0_gp,'H0','H_0')
pan_nogp.addDerived(pan_nogp.getParams().omegam_gp,'omegam','\Omega_m')
len(pan_nogp.loglikes),pan_nogp.getGelmanRubin()
```

Pantheon with GP Matern kernel

```python
pan_gp =loadMCSamples(os.path.abspath('../chains/pan_gp_matern/pan_gp_023z'),settings={'ignore_rows':0.5})
pan_gp.addDerived(pan_gp.getParams().H0_gp,'H0','H_0')
pan_gp.addDerived(pan_gp.getParams().omegam_gp,'omegam','\Omega_m')
len(pan_gp.loglikes),pan_gp.getGelmanRubin()
```

Pantheon+

```python
panp_nogp =loadMCSamples(os.path.abspath('../chains/panp_nogp/panp_nogp_calib_001z'),settings={'ignore_rows':0.9})
panp_nogp.addDerived(panp_nogp.getParams().H0_gp,'H0','H_0')
panp_nogp.addDerived(panp_nogp.getParams().omegam_gp,'omegam','\Omega_m')
len(panp_nogp.loglikes),panp_nogp.getGelmanRubin()
```

Pantheon+ with GP Matern kernel

```python
panp_gp =loadMCSamples(os.path.abspath('../chains/panp_gp_matern/panp_gp_001zhdzhel_matern'),settings={'ignore_rows':0.9})
panp_gp.addDerived(panp_gp.getParams().H0_gp,'H0','H_0')
panp_gp.addDerived(panp_gp.getParams().omegam_gp,'omegam','\Omega_m')
len(panp_gp.loglikes),panp_gp.getGelmanRubin()
```

# Constraints presented in Fig. 3

```python
gdplot = gdplt.get_subplot_plotter()
gdplot.settings.axes_fontsize=16
gdplot.settings.axes_labelsize=22
gdplot.settings.legend_fontsize=18

gdplot.triangle_plot([panp_gp,panp_nogp],['H0','omegam','d_gp','sigma_gp'],legend_loc='upper right',label_order = [1,0],legend_labels=[ 'Additional GP','Baseline analysis'],colors=['#FF2020','k'],line_args=[{'color':'#FF2020'},{'color':'k'}], filled=[True,False],alphas=[.8,1.])
gdplot.fig.suptitle('Pantheon+',y=1.02)
gdplot.fig.savefig('panp_gp_color.pdf',bbox_inches = 'tight')
```

```python
gdplot = gdplt.get_subplot_plotter()
gdplot.settings.axes_fontsize=16
gdplot.settings.axes_labelsize=22
gdplot.settings.legend_fontsize=18
gdplot.triangle_plot([pan_gp,pan_nogp],['H0','omegam','d_gp','sigma_gp'],legend_loc='upper right',label_order = [1,0],legend_labels=[ 'Additional GP','Baseline analysis'],colors=['#3820FF','k'],line_args=[{'color':'#3820FF'},{'color':'k'}], filled=[True,False],alphas=[.8,1.])
gdplot.fig.suptitle('Pantheon',y=1.02)
gdplot.fig.savefig('pan_gp_color.pdf',bbox_inches = 'tight')
```

```python
gdplot = gdplt.get_subplot_plotter(subplot_size=2)
gdplot.settings.axes_fontsize=16
gdplot.settings.axes_labelsize=22
gdplot.settings.legend_fontsize=18

gdplot.triangle_plot([panp_gp,pan_gp,pan_nogp,panp_nogp],['H0','omegam','d_gp','sigma_gp','M_gp'],legend_loc='upper right', legend_labels=['Pantheon+ GP','Pantheon GP','Pantheon','Pantheon+'],colors=[c2,c1,'k','k'],line_args=[{'color':c2},{'color':c1},{'color':'k'},{'color':'k'}],filled=[True,True,False,False],alphas=[.7,.7,1.,1.])
gdplot.fig.set_size_inches(8,8)
gdplot.fig.savefig('pan_vs_panp_nogp_color_both.pdf')
```
